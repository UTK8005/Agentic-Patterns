```python
import os
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from google.colab import userdata
```

```python
llm = ChatOpenAI(
    api_key=userdata.get('OPEN_ROUTER_KEY'),
    base_url="https://openrouter.ai/api/v1",
    model="openai/gpt-oss-20b:free",
)
```

```python
# Prompt 1 : Extracting Key Entities
prompt_extract = PromptTemplate.from_template("Extract the technical specifications from the following text:\n\n {text_input}")

# Prompt 2 : Converting extracted info to JSON
json_transform = PromptTemplate.from_template("Convert the following extracted info into a JSON object with 'Venue', 'Place' and 'Time' as keys: \n\n {specifications}")

# --- Build the Chain using LCEL ---
# The StrOutputParser() converts the LLM's message output to a simple string.
extraction_chain = prompt_extract | llm | StrOutputParser()

# The full chain passes the output of the extraction chain into the 'specifications'
# variable for the transformation prompt.
full_chain = (
   {"specifications": extraction_chain}
   | json_transform
   | llm
   | StrOutputParser()
)

```

```python
input_text="My engagement would be on September 4th 2026, Parambath Kozhikode in Miami Convention Centre"
# Execute the chain with the input text dictionary.
final_result = full_chain.invoke({"text_input": input_text})
```

```python
print("\n--- Final JSON Output ---")
print(final_result)

```
